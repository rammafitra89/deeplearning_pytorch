```python
#import module pandas
import pandas as pd
```

```python
#laod datatrain
datatrain = pd.read_csv('irisdataset/iris_train.csv')
datatrain
```

```python
#change string value to numeric
datatrain.loc[datatrain['species']=='Iris-setosa', 'species']=0
datatrain.loc[datatrain['species']=='Iris-versicolor', 'species']=1
datatrain.loc[datatrain['species']=='Iris-virginica', 'species']=2
datatrain = datatrain.apply(pd.to_numeric) # memastikan semuanya numeric
datatrain
```

```python
#change dataframe to array
datatrain_array = datatrain.values
```

```python
datatrain_array
```

```python
#split x and y (feature and target)
xtrain = datatrain_array[:,:4]
ytrain = datatrain_array[:,4]
```

```python
"""
SECTION 2 : Build and Train Model
Multilayer perceptron model, with one hidden layer.
input layer : 4 neuron, represents the feature of Iris
hidden layer : 3 neuron, activation using ReLU
output layer : 3 neuron, represents the class of Iris
optimizer = stochastic gradient descent with no batch-size
loss function = categorical cross entropy
learning rate = 0.01
epoch = 50
"""
```

```python
#import module pythorch
import torch
import torch.nn as nn
import torch.nn.functional as F
torch.manual_seed(1234)
```

```python
#Hyperparameter
hl = 3
lr = 0.01
num_epoch = 50
```

```python
# build model
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.fc1 = nn.Linear(4,3) #bobot antara input dan hidden layer
        self.fc2 = nn.Linear(3,3) #bobot antara hidden dan output layer
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

ourmodel = Net() #menginisalisasi objek
```

```python
# algoritma optimasi -> Stochastic Gradient Descent (SGD) dengan learning rate sebesar 0.01
optimizer = torch.optim.SGD(ourmodel.parameters(), lr=lr)
```

```python
# metode untuk menghitung error -> Cross-Entropy
criterion = nn.CrossEntropyLoss()
```

```python
# this is part of training model
for epoch in range (num_epoch):
    # 1. ubah semua variabel ke tipe tensor
    X = torch.Tensor(xtrain).float()
    Y = torch.Tensor(ytrain).long()
    #2. Inputkan data ke model untuk mendapatkan output model
    out = ourmodel(X)
    #3. Hitung eror antara output dengan label
    loss = criterion(out, Y)
    #4. Bersihkan riwayat gradient sebelumnya menggunakan zero_grad()
    optimizer.zero_grad() 
    #5. Hitung gradient eror dengan backward()
    loss.backward()  
    #6.Update bobot menggunakan algoritma optimasi
    optimizer.step() 
    acc = 100 * torch.sum(Y==torch.max(out.data, 1)[1]).double() / len(Y)
    print ('Epoch [%d/%d] Loss: %.4f' %(epoch+1, 50, loss.item()))
   
```

```python
# ambil nilai terbesar setiap kelasnya sebagai prediksi
prediksi = torch.max(out.data, 1)[1] 
# bandingkan dengan label, hitung rata-rata persentase
akurasi = 100 * torch.sum(Y==prediksi).double() / len(Y)
print(akurasi)
```
